```python
# Perceptron for Even-Odd Classification (ASCII input)

# Step 1: Activation Function
def step(x):
    if x >= 0:
        return 1
    else:
        return 0


# Step 2: Training Data (0–9)
X = [i % 2 for i in range(10)]   
Y = [i % 2 for i in range(10)]  

# Step 3: Initialize weights and bias
w = 0
b = 0


# Step 4: Training with epochs
epochs = 10

for epoch in range(epochs):
    for i in range(10):
        x = X[i]
        y = Y[i]
        
        # Prediction
        y_hat = step(w * x + b)
        
        # Update rule (Perceptron Learning)
        w = w + (y - y_hat) * x
        b = b + (y - y_hat)


# Step 5: Testing with ASCII input
char = input("Enter a digit (0-9): ")

if char not in "0123456789":
    print("Invalid input")
else:
    # Convert ASCII to number
    num = ord(char) - 48
    
    # Feature extraction
    x = num % 2
    
    # Prediction
    y_pred = step(w * x + b)
    
    # Output
    if y_pred == 0:
        print("Even")
    else:
        print("Odd")


# Step 6: Show learned parameters 
print("Learned weight:", w)
print("Learned bias:", b)
```
