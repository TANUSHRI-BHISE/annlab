```python
import numpy as np
# Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(a):
    return a * (1 - a)   # a = sigmoid(z)
    
# Training Data
X = np.array([[1, 0, 1]])   # Input (1×3)
Y = np.array([[1]])         # Target (1×1)
# Initialize Parameters
w = np.random.rand(3, 1)    # Weights (3×1)
b = 0.0                     # Bias (scalar)
lr = 0.5                    # Learning rate
# Training Loop
for epoch in range(10):

    #  Forward Propagation
    z = np.dot(X, w) + b
    output = sigmoid(z)

    # Loss (MSE)
    loss = 0.5 * (Y - output) ** 2

    #  Backpropagation (Chain Rule)
    d_loss = (output - Y)                     # dL/da
    d_output = d_loss * sigmoid_derivative(output)   # dL/dz (delta)


    # Chain rule:
# dL/dw = dL/da * da/dz * dz/dw
# dL/db = dL/da * da/dz * dz/db
    
    # Gradient Descent Update
    w -= lr * np.dot(X.T, d_output)          # weight update
    b -= lr * d_output.item()               # bias update (scalar)

    print(f"Epoch {epoch+1}: Output={output}, Loss={loss}")


# -----------------------------
# Final Output
# -----------------------------
print("\nFinal Output:", output)
print("Final Weights:\n", w)
print("Final Bias:", b)
```
